```python
import pandas as pd
dub1 = pd.read_csv("tweeter_dub1.csv", delimiter=";")
dub2 = pd.read_csv("tweeter_dub2.csv", delimiter=";")

positive_words = '''
хороший
спасибо
радостный
вкусный
забавный
милый
уютный
веселый
мелодичный
свободный
ласковый
впечатляющий
талантливый
'''

negative_words = '''
ненавистный
безответственный
дикий
наказание
жестокость
убийство
беда
виновный
варварство
грустно
печально
плохой
фейк
'''

def rm_empty(lst):
    return [ x for x in lst.split("\n") if x ]

# positive_words = positive_words.split("\n")
# positive_words = list(filter(None, positive_words))
positive_words = rm_empty(positive_words)
negative_words = rm_empty(negative_words)
```

```python
dub1
```

```python
dub2
```

```python
def get_tonality(text):
    tonality = 0
    words = text.split()
    for w in words:
        if w in positive_words:
            tonality += 1
        if w in negative_words:
            tonality -= 1
    return tonality
```

# 1. Определить, на каком из дубов в среднем более добрые посты;

```python
dub1["тональность"] = dub1["текст поста"].apply(lambda x: get_tonality(x))
dub2["тональность"] = dub2["текст поста"].apply(lambda x: get_tonality(x))
```

# 2. Для дуба, который в среднем менее добрый, определить количество дней, в которые его посты в среднем добрее

```python
df1 = dub1[["дата поста","тональность"]].groupby(["дата поста"]).agg("sum")
df2 = dub2[["дата поста","тональность"]].groupby(["дата поста"]).agg("sum")
#dfc = pd.concat([df1, df2], axis=1,keys=["dub1","dub2"])
dfc = df1.join(df2, lsuffix='_дуб1', rsuffix='_дуб2')
dfc["is_more"] = dfc["тональность_дуб1"] - dfc["тональность_дуб2"]
days_qty = len(dfc.query("is_more > 0"))
days_qty
```
